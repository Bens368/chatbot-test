```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin, urlparse

visited = set()

def sauvegarder_contenu(url, texte):
    # Sauvegarde le contenu de la page dans un fichier texte
    with open("site_content.txt", "a", encoding="utf-8") as f:
        f.write(f"URL: {url}\n")
        f.write(texte)
        f.write("\n" + "="*80 + "\n")  # Séparateur entre les pages

def recuperer_contenu_textuel(url):
    try:
        response = requests.get(url)
        response.raise_for_status()  # Lève une exception pour les erreurs HTTP
        soup = BeautifulSoup(response.text, 'html.parser')
        # Extraire le texte brut de la page
        texte = soup.get_text(separator=' ', strip=True)
        return texte, soup
    except requests.RequestException as e:
        print(f"Erreur lors de la requête HTTP sur {url} : {e}")
        return None, None

def est_interne(url, domaine):
    # Vérifie si l'url appartient au même domaine
    return urlparse(url).netloc == domaine

def crawler(url, domaine, profondeur=0, max_profondeur=2):
    if url in visited or profondeur > max_profondeur:
        return
    visited.add(url)
    print(f"Exploration de {url}")
    
    texte, soup = recuperer_contenu_textuel(url)
    if texte:
        # Sauvegarder l'intégralité du contenu textuel dans un fichier
        sauvegarder_contenu(url, texte)
        
    if soup:
        for lien in soup.find_all('a', href=True):
            next_url = urljoin(url, lien['href'])
            # Filtrer les liens pour ne crawler que ceux du même domaine
            if est_interne(next_url, domaine):
                crawler(next_url, domaine, profondeur + 1, max_profondeur)

# Exemple d'utilisation
site_url = "https://www.dr-benadiba.ch/"
domaine = urlparse(site_url).netloc
crawler(site_url, domaine)
```
